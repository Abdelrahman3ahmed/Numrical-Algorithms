# Secant Method

```python
def Secant_Method(xk , x_prevoius , Max_iter , epsilon):
    for i in range(Max_iter):
        if(np.abs(xk-x_prevoius) < epsilon):
            print("Stopped at iter ", i+1)
            break
        f_dash_xk = (Secant_equatin(xk) - Secant_equatin(x_prevoius))/(xk-x_prevoius)
        xk_next = xk - (Secant_equatin(xk) / f_dash_xk)
        x_prevoius = xk
        xk = xk_next 
    return xk_next , Secant_equatin(xk_next)
```

```python
def Secant_equatin(x):
    f_X = x**3 - x**2 -15*x +1
    return f_X
```

```python
root , f_root = Secant_Method(4.5 , 5.5, 20 , 1e-8)
print("r",root , "\nf(r1)" ,f_root )
```
