# Secant Method

```python
def Secant_Method(xk , x_prevoius , Max_iter , epsilon):
    for i in range(Max_iter):
        if(np.abs(xk-x_prevoius) < epsilon):
            print("Stopped at iter ", i+1)
            break
        f_dash_xk = (Secant_equatin(xk) - Secant_equatin(x_prevoius))/(xk-x_prevoius)
        xk_next = xk - (Secant_equatin(xk) / f_dash_xk)
        x_prevoius = xk
        xk = xk_next 
    return xk_next , Secant_equatin(xk_next)
```

```python
def Secant_equatin(x):
    f_X = x**3 - x**2 -15*x +1
    return f_X
```

```python
root , f_root = Secant_Method(4.5 , 5.5, 20 , 1e-8)
print("r",root , "\nf(r1)" ,f_root )
```

# Gradient Method (Single)

```python
def Gradient_single(xk , Max_iter , epsilon , alpha):
    for i in range(Max_iter):
        xk = xk - (alpha * f_dash_x(xk))
        if(np.abs(f_dash_x(xk)) < epsilon):
            print("Stopped at iter ", i+1)
            break
            
    return xk , f_x(xk)
```

# Gradient Method (MultiVariable)

```python
def Gradient_Multi(xk , Max_iter , epsilon , alpha):
    for i in range(Max_iter):
        xk = xk - (alpha * f_dash_x(xk))
        if(np.linalg.norm(f_dash_x(xk)) < epsilon):
            print("Stopped at iter ", i+1)
            break
            
    return xk , f_x(xk)
```

# BFGS

```python
def BFGS(B_mat , xk , x_prevois , Max_iter , epsilon , alpha):

    for i in range(Max_iter):
        Delta_x = xk - x_prevois
        y = f_dash_x_xpre(xk) - f_dash_x_xpre(x_prevois)
        f_com = np.eye(2) - ((Delta_x @ y.T) / (y.T @ Delta_x))
        s_com = np.eye(2) - (y @ Delta_x.T) / (y.T @ Delta_x)
        th_com = (Delta_x @ Delta_x.T) / (y.T @ Delta_x)
        B_mat = (f_com @ B_mat) @s_com + th_com
        x_prevois = xk
        xk = xk - (alpha * B_mat.dot( f_dash_x_xpre(xk)))
        
        if(np.linalg.norm(f_dash_x_xpre(xk)) < epsilon):
            print("Stopped at iter ", i+1)
            break
    return B_mat , xk
```
